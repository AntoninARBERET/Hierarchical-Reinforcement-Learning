## Plot functions

```python
#data is a list of tuple (x, y, label)
def print_refin(data, x_unit):
    i=0
    pts = ['C0o','C1o','C2o']
    for (x,y,lab) in data:
        plt.step(x, y, where='post', label=lab)
        plt.plot(x, y , pts[i], alpha=0.5)
        i+=1
    
    plt.legend(title='Agents:')
    plt.xlabel(x_unit)
    plt.ylabel("Correct refinements")
    plt.title('Correct refinements through time')
    return plt
```

```python
def print_exp_iter(data_path, experiment_id, percent=100):
    folder_path = data_path+experiment_id
    if(not path.exists(folder_path)):
        print("Folder not found")
        return
    types = ["Random","Local","Global"]
    data=[]
    l=0
    for t in types:
        if(path.exists(folder_path+"/"+t)):
            if(len(np.loadtxt(folder_path+"/"+t+"/iterations"))>l):
                l=len(np.loadtxt(folder_path+"/"+t+"/iterations"))
    for t in types:
        if(path.exists(folder_path+"/"+t)):
            x = np.loadtxt(folder_path+"/"+t+"/iterations")
            y = np.loadtxt(folder_path+"/"+t+"/refinements")
            limit=(l*percent)//100
            label = t
            if(limit>len(x)):
                limit=len(x)
                
            data.append((x[:limit],y[:limit],label))
    return print_refin(data, "iterations")
```

```python
def print_exp_time(data_path, experiment_id, percent=100):
    folder_path = data_path+experiment_id
    if(not path.exists(folder_path)):
        print("Folder not found")
        return
    types = ["Random","Local","Global"]
    data=[]
    l=0
    for t in types:
        if(path.exists(folder_path+"/"+t)):
            if(len(np.loadtxt(folder_path+"/"+t+"/time"))>l):
                l=len(np.loadtxt(folder_path+"/"+t+"/time"))
    for t in types:
        if(path.exists(folder_path+"/"+t)):
            x = np.loadtxt(folder_path+"/"+t+"/time")
            y = np.loadtxt(folder_path+"/"+t+"/refinements")
            limit=(l*percent)//100
            label = t
            if(limit>len(x)):
                limit=len(x)
                
            data.append((x[:limit],y[:limit],label))
    return print_refin(data, "time (s)")
```
